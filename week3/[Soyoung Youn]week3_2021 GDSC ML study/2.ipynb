```python
import torch

# set up simple graph relating x, y and z
x = torch.tensor(3.5, requires_grad=True)
y = x*x
z = 2*y + 3
```

```python
# work out gradients
z.backward()
```

```python
# what is gradient at x = 3.5
print(x.grad)   # dz/dx is in x.grad(not dy/dx)
```
