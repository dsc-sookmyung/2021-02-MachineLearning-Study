```python
import torch

# simple pytorch tensor
x = torch.tensor(3.5)
print(x)
```

```python
# simple arithmetic with tensors
y = x + 3
print(y)
```

```python
# pytorch tensor with gradient enabled
x = torch.tensor(3.5, requires_grad=True)
print(x)
```

```python
# y is defined as a function of x
y = (x-1) * (x-2) * (x-3)
print(y)
```

```python
# work out of gradient
y.backward()
```

```python
# what is gradient at x = 3.5
x.grad
```

```python
# example of general python variable's use
x = 3.5
y = x*x + 2
print(x, y)
```
