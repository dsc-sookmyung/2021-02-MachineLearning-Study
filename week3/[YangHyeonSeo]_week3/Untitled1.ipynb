```python
x = torch.tensor(3.5, requires_grad=True)
print(x)
```

```python
y = (x-1) * (x-2) * (x-3)
print(y)
```

```python
y.backward()
```

```python
x.grad
```

```python
print(x, y, x.grad)
```

```python
#계산그래프
x = torch.tensor(3.5, requires_grad=True)
y = x*x
z = 2*y + 3
```

```python
z.backward()
```

```python
print(x.grad)
```

```python
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)

x = 2 * a + 3 * b
y = 5*a*a + 3*b*b*b
z = 2*x + 3*y
```

```python
z.backward()

a.grad
```
